# Note 
1. Youc can store the params in its own variable and then apply the fstring

````python
# Testing it out 

import requests as rq

def get_prayer_times():
    
    # Breaking down all the elements for easier management and readaibility
    lon = -0.127758
    lat = 51.507351
    date = "15-11-2024"
    endpoint = "http://api.aladhan.com/v1/timings/"
    
    # Constructing the URL 
    url = f"{endpoint}{date}?latitude={lat}&longitude={lon}"
    
    # Actual code to get the response
    response = rq.get(url).json()
    pprint(response, expand_all="true")

get_prayer_times()
````

````python
# Above code with exception blocks 
import requests as rq

def get_prayer_times():
    lon = -0.127758
    lat = 51.507351
    date = "2024-11-15"
    endpoint = "http://api.aladhan.com/v1/timings/"
    
    url = f"{endpoint}{date}?latitude={lat}&longitude={lon}"
    
    try:
        response = rq.get(url)
        response.raise_for_status()
        data = response.json()
        pprint(data, expand_all=True)
    except rq.exceptions.RequestException as e:
        print(f"An error occurred: {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

get_prayer_times()
````
